Recursion:

```python
def factorial(x):
    if x==1:
        return 1
    else:
        return(x*factorial(x-1))   #factorial is a recursive function as it call itself
num=int(input("Enter a number:"))
print("The factorial of", num, "is", factorial(num))



    
    
```
