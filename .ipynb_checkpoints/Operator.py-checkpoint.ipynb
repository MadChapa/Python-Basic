Arithemetic Operator

```python
x=15
y=4

print("x+y:", x+y)

print("x-y:", x-y)

print("x*y:", x*y)

print("x/y:", x/y)

print("x%y:", x%y)

print("x//y:", x//y)

print("x**y:", x**y)
```

Comparision Operator:

```python
x=10
y=12
print('x>y is',x>y)

print('x<y is',x<y)

print('x==y is',x==y)

print('x!=y is',x!=y)

print('x>=y is',x>=y)

print('x<=y is',x<=y)
```

Logical operator:

```python
x=True
y=False
print('x and y is',x and y)

print('x or y is', x or y)

print('not x is',not x)
```

```python
Bitwise Operator
```

```python
a=10<<2
print(a)

a=10<<3
print(a)
```

Identity Operator:

```python
x1=5
y1=5
x2='hello'
y2='hello'
x3=[1,2,3]
y3=[1,2,3]
print(x1 is not y1)

print(x2 is y2)

print(x3 is y3)
```

Membership Operator

```python
x='Hello world'
y={1:'a', 2:'b'}
print('H' in x)

print('hello' not in x)

print(1 in y)

print('a' in y)
```
