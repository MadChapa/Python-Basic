if-elif-else:

```python
num=int(input("Enter a number:"))
if num>0:
    print(num,"is a postive number")
elif num==0:
    print(num,"is neither positive nor negative")
else:
    print(num,"is a negative number")

    
```

Nested-if:

```python
num=int(input("Enter a number:"))
if num>0:
    if num==0:
        print(num,"is neither positive nor negative")
    else:
        print(num,"is a postive number")
else:
    print(num,"is a negative number")



        
```
