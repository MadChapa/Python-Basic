Implicit type conversion:
 automatically lower data type lai higher data type mah convert garxa
    in the below program integer automatically float mah convert vaeko xa!!

```python
num_int=123
num_flo=1.23
num_new=num_int+num_flo
print("Value of new_num=",num_new)
```

Explicit Type casting

```python
num_int=123
num_str="456"
print("Data type of num_int:",type(num_int))

print("Data type of num_str before type casting :",type(num_str))

num_str=int(num_str)
print("Data type of num_str after type casting :",type(num_str))

num_sum=num_int+num_str
print("Sum :",num_sum)

```
