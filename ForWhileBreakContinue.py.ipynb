for loop:

```python
numbers=[2,3,4,5,6,7,8]
sum=0  #variable to store the sum
for val in numbers:
    sum=sum+val
print("The sum is",sum)
```

While Loop:

```python
num=int(input("Enter a number:"))
sum=0   #initialize sum and counter
i=1
while i<=num:
    sum=sum+i
    i=i+1
print("The sum is:",sum)
```

Break Statement;

```python
for val in "string":
    if val == "i":
        break
    print(val)
print("the end")
```

Continue Statement:

```python
for val in "String":
    if val=="i":
        continue
    print(val)
print("The end")
        
```
